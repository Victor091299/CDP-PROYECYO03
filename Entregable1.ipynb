OBTENIENDO MEDIA A TRAVEZ DE UNA FUNCION

```python
#CREANDO LA FUNCION LLAMADA MEDIA 
def media(x):
  '''
  NOS RETORNAR√Å EL PROMEDIO DE CADA VARIBALE
  '''
  return stats.mean(x)
```

```python
#APLICANDO LA FUNCION MEDIA A LAS VARIABLE NUMERICAS DE MI DATAFRAME
datos_nutricion[var_num].apply(media,axis=0).round(2)
```

OBTENIENDO LA MEDIANA A TRAVEZ DE UNA FUNCION 

```python
#CREANDO LA FUNCION LLAMADA MEDIANA 
def mediana(x):
  '''
  NOS DAR√Å EL VALOR QUE OCUPA LA POSICION CENTRAL DE CADA VARIABLE
  '''
  return stats.median(x)
```

```python
#APLICANDO LA FUNCION MEDIANA A LAS VARIABLES NUMERICAS DE MI DATAFRAME
datos_nutricion[var_num].apply(mediana,axis=0).round(2)
```

OBTENIENDO LA MODA A TRAVEZ DE UNA FUNCION

```python
#CREANDO LA FUNCION LLAMADA MODA
#DE COLLECTIONS IMPORTAMOS LA LIBRERIA COUNTER QUE NOS AYUDARA A CONTAR LA CANTIDAD DE ELEMENTOS DE CADA VARIABLE
from collections import Counter

def moda(x):
  '''
  NOS RETORNAR√Å EL VALOR QUE MAS SE REPITE POR VARIABLE Y EL NUMERO DE VECES QUE SE REPITE
  '''
  c=Counter(x) #LA CANTIDAD DE ELEMENTOS DE CADA VARIABLE SE ALAMACENAR√Å EN EL IDENTIFICADOR "C"
  return c.most_common(1) #A "C" SE LE APLICAR√Å LA FUNCION MOS_COMON QUE NOS RETORNAR√Å EL VALOR MAS REPETIDO Y LA CANTIDAD DE VECES QUE SE REPITE.
```

```python
#APLICANDO LA FUNCION MODA A LAS VARIABLES NUMERICAS DE MI DATAFRAME
datos_nutricion[var_num].apply(moda,axis=0)
```

OBTENIENDO EL CENTRO DE AMPLITUD A TRAVEZ DE UNA FUNCION 

```python
#CREANDO LA FUNCION LLAMADA CENTRO DE AMPLITUD
def centro_apli(x):
  '''
  COMO RESULTADO NOS DAR√Å EL VALOR MEDIO ENTRE EL MINIMO Y MAXIMO 
  '''
  return (max(x) - min(x))/2
```

```python
#APLICANDO LA FUNCION CENTRO_AMPLI A LAS VARIABLES NUMERICAS DE MI DATAFRAME
datos_nutricion[var_num].apply(centro_apli,axis=0).round(2)
```

OBTENIENDO LA MEDIA GEOMETRICA A TRAVEZ DE UNA FUNCION 

```python
#CREANDO LA FUNCION LLAMADA MEDIA GEOMETRICA
def med_geome(iterable):
  '''
  REALIZA EL CALCULO DE LA MEDIA GEOMETRICA

  MULTIPLICAR√Å LOS ELEMENTOS DE CADA VARIABLE PARA LUEGO CALCULAR LA RAIZ CON EL EXPONENTE N(n¬∞ DE ELEMENTO DE LA VARIABLE)
  '''
  a=np.log(iterable)
  return "Media Geometrica",np.exp(a.sum()/len(a)).round(2)
```

```python
#APLICANDO LA FUNCION MED_GEOME
datos_nutricion[var_num].apply(med_geome,axis=0)
```

OBTENIENDO LA MEDIA ARMONICA A TRAVEZ DE UNA FUNCION

```python
#CREANDO LA FUNCION LLAMADA MEDIA ARMONICA
def media_armonica(x):
  return "Media Armonica", round(stats.harmonic_mean(x),2)
```

MEDIDAS DE DISPERSION

```python
#RANGO
#CREANDO LA FUNCION LLAMADA RANGO

def rango(x):
  '''
  RESTAR√Å EL MINIMO Y MAXIMO VALOR
  DE UN CONJUNTO DE ELEMENTOS

  COMO RESULTADO NOS DAR√Å EL RANGO
  '''
  min=np.min(x)
  max=np.max(x)
  r= max - min
  return min,max,r
```

```python
#APLICANDO LA FUNCION RANGO A LAS VARIABLES NUMERICAS DE MI DATAFRAME
p=datos_nutricion[var_num].apply(rango,axis=0).round(2).T
p.columns=["Minimo","Maximo","Rango"]
p
```

```python
#CUARTILES
#CREANDO FUNCION LLAMADA CUARTILES
def cuartiles(x: pd.Series):
    '''
    Nos retornar√° como resultado los cuariles 
    25,50,75 y los valores min y max 
    '''
    Q1 = x.quantile([.25]).values[0]
    Q2 = x.quantile([.50]).values[0]
    Q3 = x.quantile([.75]).values[0]
    min = np.min(x)
    max = np.max(x)
    return min, Q1, Q2, Q3, max
```

```python
#APLICANDO LA FUNCION LLAMADA CUARTILES
j = datos_nutricion[var_num].apply(cuartiles, axis=0).round(2).T
j.columns = ["M√≠nimo","P25% - Q1","P50% - Q2", "P75% - Q3", "M√°ximo"]
j
```

```python
#RANGO INTERCUARTIL
#CREANDO LA FUNCION LLAMADA RANGO INTERCUARTIL
def rango_intercuartil(x: pd.Series):
    '''
    NOS RETORNAR√Å EL CUARTIL 25,75 Y EL 
    RANGO INTERCUARTIL
    '''
    Q1 = x.quantile([.25]).values[0]
    Q3 = x.quantile([.75]).values[0]

    return Q1, Q3 - Q1, Q3
```

```python
#APLICANDO LA FUNCION LLAMADA RANGO INTERCUARTIL
o=datos_nutricion[var_num].apply(rango_intercuartil, axis=0).T
o.columns = ["Q1","RIC","Q3"]
o
```

```python
#COEFICIENTE DE VARIACION 
def coeficiente_variacion(x):
  m = stats.mean(x)
  sd = np.std(x)
  return sd/m*100
```

```python
#OBESERVANDO LA CURTOSIS 
from scipy.stats import kurtosis

def kurtosis_data(x):
  curtosis =  kurtosis(x, fisher=True)

  if curtosis >0:
    return ('Es Leptoc√∫rtica-->', round(curtosis,2 ))
  elif curtosis ==0: # 0.03
    return  ('Es Mesoc√∫rtica-->', round(curtosis,2 ))
  else:
    return  ('Es Platic√∫rtica-->', round(curtosis,2 ))
```

```python
datos_nutricion[var_num].apply(kurtosis_data, axis=0).T
```

CORRELACION DE PEARSON

```python
r = datos_nutricion[var_num].corr(method='pearson').round(3) # param√©trico -> X y Y: distribuci√≥n normal
r
```

```python
mask = np.triu(np.ones_like(r, dtype=bool )) #la m√°scara para el tri√°ngulo superior del gr√°fico
f, ax = plt.subplots(figsize = (10,10)) # la base para el gr√°fico
g = sns.heatmap(r, mask = mask, cmap='plasma', vmax=1, center = 0, 
            square = True, linewidths=.1, annot=True)

g.set_title("Gr√°fico N¬∫ XX: Gr√°fico de la correlaci√≥n r de pearson")
```

CORRELACION RHO DE SPEARMAN

```python
Rho = datos_nutricion[var_num].corr(method='spearman').round(3) # no param√©trico
Rho
```

```python
mask = np.triu(np.ones_like(Rho, dtype=bool )) #la m√°scara para el tri√°ngulo superior del gr√°fico
f, ax = plt.subplots(figsize = (10,10)) # la base para el gr√°fico
g = sns.heatmap(Rho, mask = mask, cmap='viridis', vmax=1, center = 0, 
            square = True, linewidths=.1, annot=True)

g.set_title("Gr√°fico N¬∫ XX: Gr√°fico de la correlaci√≥n Rho de Spearman")
```
