Prueba de normalidad

---


PLANTEAMIENTO DE LAS HIPóTESIS FINALIDADES :Verificar si los datos estan normalmente distribuidos o no . Si llegasen a estar normalmente distribuidos haríamos una prueba parametrica y si fuera lo contrario se haría una prueba no parametrica.

Ho: Los datos estan normalmente distribuidos

Ha: Los datos no estan normalmente distribuidos

Nivel de significancia = 5% (0.05)

```python
#import matplotlib.pyplot as plt
from statsmodels.graphics.gofplots import qqplot #UTILIZANDO EL GRAFICO qqplot
from scipy.stats import shapiro # n <= 50
from scipy.stats import kstest  # n > 50
from scipy.stats import zscore
```

```python
#PRUEBA DE NORMALIDAD:TALLA
qqplot(datos_nutricion_m['talla'], line='s')#REALIZANDO EL GRAFICO DE qqplot CON NUESTRA VARIBALE talla
                                       #line=s nos sirve para colocar la linea roja diagonal
plt.show()
```

```python
# Prueba de Shapiro-Wilk (n < =50)
stat, p = shapiro(datos_nutricion_m['talla'])

print('Estadisticos=%.3f, p-value=%.3f' % (stat, p))

#VEMOS QUE NUESTRO P VALUE ES < AL NIVEL DE SIGNIFICANCIA.POR LO TANTO 
#EN ESTE CASO SOLO NOS QUEDA ACEPTAR LA HIPOTESIS ALTERNA(Ha) Y RECHAZAR LA HIPOTESIS
#NULA(Ho).
```

```python
#Prueba de kstest (n > 50)
stat2, p2 = kstest(datos_nutricion_m['talla'], 'norm') 

print('Estadisticos=%.3f, p-value=%.3f' % (stat2, p2))

#VEMOS QUE NUESTRO P VALUE ES < AL NIVEL DE SIGNIFICANCIA.
#ENTONCES SOLO NOS QUEDA ACEPTAR LA HIPOTESIS ALTERNA(Ha)
```

```python
#DECISION: CON UN NIVEL DE SIGNIFICANCIA DEL 5%(0.05)AFIRMAMOS QUE NUESTRA VARIABLE TALLA NO ESTA NORMALMENTE DISTRIBUIDA.POR LO TANTO 
#MAS ADELANTE SE LE TENDRÁ QUE HACER UNA PRUEBA NO PARAMETRICA.
```

 Normalizando variables / transformadas

 
normalización = ( x – min(x) ) / ( max(x) – min(x) )

Los valores normalizados para todos los demás valores del conjunto de datos estarán entre 0 y 1.

```python
#NORMALIZANDO CON MIN-MAX
# Normalizamos las variables numéricas

from sklearn.preprocessing import MinMaxScaler#DENTRO DE LA LIB DE SKLEARN EN EL MODULO DE PREPROCESING ESTA LA FUNCION MINMAXSCALER

scaler = MinMaxScaler()
datos_cuanti2_norm = scaler.fit_transform(datos_cuanti2) #APLICANDO A NUESTRA VARIABLES CUALITATIVAS "data_riesgo_num2"
datos_cuanti2_norm

#NOS DA COMO RESULTADO UN ARRAY PERO TENEMOS QUE TRANSFORMARLO A UN DTF
```

```python
datos_cuanti2_norm = pd.DataFrame(
    data=datos_cuanti2_norm,
    columns=["talla","edad","peso_kg","circun_cuello","IMC"])
datos_cuanti2_norm.info()
```

```python
qqplot(datos_cuanti2_norm['talla'] , line='s')
plt.show()
```

```python
# Test de Kolmogorov-Smirnov
#Cuando n > 50

stat2, p2 = kstest(datos_cuanti2_norm['talla'] , 'norm')
print('Estadisticos=%.3f, p-value=%.3f' % (stat2, p2))
```
